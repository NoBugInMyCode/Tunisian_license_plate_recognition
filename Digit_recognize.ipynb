## Import libraries

```python
import numpy as np
import os
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

## Create `datagen` to modify pictures
#### ie. rotate pictures, change picture size
#### Make the model adaptable to complex situations

```python
datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2
)
```

## Load training dataset

```python
train_data = datagen.flow_from_directory(
    'LP_digit/traning_data_set',
    target_size=(64, 64),
    batch_size=32,
    class_mode='categorical',
    subset='training'
)
```

## Load testing dataset

```python
validation_data = datagen.flow_from_directory(
    'LP_digit/traning_data_set',
    target_size=(64, 64),
    batch_size=32,
    class_mode='categorical',
    subset='validation'
)
```

## Create model(CNN)

```python
model = Sequential()

model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 3)))
model.add(MaxPooling2D(2, 2))

model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(2, 2))

model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D(2, 2))


model.add(Flatten())

model.add(Dense(128, activation='relu'))
model.add(Dense(11, activation='softmax'))
```

## Train model

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
history = model.fit(train_data, epochs=150, validation_data=validation_data)
```

## Print accuracy

```python
train_accuracy = history.history['accuracy'][-1]
val_accuracy = history.history['val_accuracy'][-1]

print(f"Training Accuracy after last epoch: {train_accuracy*100:.2f}%")
print(f"Validation Accuracy after last epoch: {val_accuracy*100:.2f}%")
```

## Save model

```python
model.save('LP_digit.h5')
```
